# Github

## 준비 단계

1. Github 가입하기 : [https://github.com](https://github.com/)
2. Git 다운로드 [https://git-scm.com/](https://git-scm.com/) 
3. Git bash 실행하기
4. Git bash 환경설정
    1. 유저 이름 설정하기 : **git config --global [user.name](http://user.name) “설정할 이름”**
    2. 유저 이메일 설정하기 : **git config --global [user.email](http://user.email) “Github가입한 이메일”**
    3. 확인하기 : **git config --list** 명령어를 입력하고 a, b에서 입력한 정보 확인

## Github 업로드하기

1. 초기화 하기
    
    **git init** 명령어를 사용하여 초기화
    
     ※ 처음 업로드를 진행할 때 진행함(추가 업로드를 할 경우 진행하지 않음)
    
2. 업로드 할 파일 선택
    
    **git add .** 명령어를 사용하면 모든 파일을 선택
    
     → 모든 파일이 아닌 특정 파일을 선택하고 싶은 경우 . 대신 파일명을 작성 ex) **git add test.py**
    
3. 히스토리 만들기
    
    **git commit -m “히스토리”** 명령어를 사용한다. “히스토리는” 자신이 작성하고 싶은 메시지라고 생각하면 쉽게 이해할 수 있다. 예를들면 팀프로젝트를 진행하며 누가 커밋 했는지 표시하고자 하면 **git commit -m “geunyoung commit”** 이런 형식으로 작성하면 
    
    geunyoung commit으로 Github에 표시되어 간략한 설명을 작성하면 좋을 것 같다.
    
4. Github repository와 연결하기
    
    Github에서 repository를 처음 만들면
    
    #### ...or creat a new repositoy in the command line
    
    을 볼 수 있는데 커맨드라인 중 아래에서 2번째 명령어를 그대로 복사 후 붙여넣기 하면 된다.
    
    **git remote add origin [https://github.com/Kim-Geunyoung/rmsdud.git](https://github.com/Kim-Geunyoung/rmsdud.git)** 이 명령어는 해당repository로 내 프로젝트 파일을 보내기 위해 연결한다. 라고 생각하면 이해하기 쉬울 것이다.
    
    ※ 위 명령어는 repository마다 다르니까 위 사진에 보이는 화면에서 복사 붙여넣기로 사용하는걸 추천하고, 해당 작업도 최초 1회만 하면 되니까 추가 업로드를 할 경우에는 생략
    
5. Github 업로드
    
    **git push origin branch이름** 명령어로 업로드하기
    
    ※ 위 명령어의 branch이름은 기본적으론 master로 설정되고 협업을 할 경우 branch를 생성하는데 그때는 branch 명을 입력해주면 된다!
    
6. 추가 업로드
    
    2, 3, 5번의 과정만 실행하여 추가 업로드 가능
    

## 협업

1. Github에서 코드 다운받기
    
    명령 프롬프트(cmd)에서 **git clone 주소** 명령어를 입력한다. 이때 주소는 SSH로 진행하였다.
    

1. branch 만들기
    
    **git checkout -b branch 이름** 명령어를 사용해서 만들 수 있다.
    
    

1. 업로드
    
    Github 업로드 하는 방법 그대로 진행하면 됨(업로드 파일 선택, 히스토리 만들기, 업로드)
    
    ※ git push origin master로 업로드 하면 바로 적용되지만 master가 아닌 branch명으로 업로드를 진행하게 되면 Github 사이트에서 따로 승인 해줘야 업로드가 완료됨
    
2. 코드 최신화
    
    누군가가 소스코드를 업로드 해서 Github에 파일이 변경이 된다면 변경된 소스코드로 최신화를 해야한다. 그렇지 않으면 나중에 서로 다른 소스코드로 개발하며 오류가 발생하고 수정하기 매우 힘들다…
    
    **git pull origin master** 명령어를 사용하여 최신화 할 수 있다.
