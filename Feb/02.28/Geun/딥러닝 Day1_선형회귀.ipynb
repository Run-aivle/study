## 선형회귀
> 한 개 이상의 독립변수 x와 x값에 의해 종속적으로 값이 변하는 y의 선형 상관 관계를 모델링하는 회귀분석 방법  
**1. 단순 선형 회귀 분석(Simple Linear Regression Analysis)**  
$ y = wx + b $  
**2. 다중 선형 회귀 분석(Multiple Linear Regression Analysis)**  
$ y = w_1x_1 + w_2x_2 + ... + w_nx_n + b $

* 위 선형회귀 내용 추가하기

## 실습

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
x = np.array(range(30))
y = x * 3
print(f'x = {x}')
print(f'y = {y}')
```

```python
x.shape, y.shape
```

```python
# 선형회귀 Sequential API 방식
# 세션 초기화
keras.backend.clear_session()

# 모델 선언
model = keras.models.Sequential()

# 모델 레이어 쌓기
model.add(keras.layers.Input(shape=(1,)))
model.add(keras.layers.Dense(1, activation='linear'))
```

```python
# 모델 컴파일
model.compile(loss='mse', optimizer='adam')
```

```python
# 모델 학습
model.fit(x, y, epochs=100, verbose=1)
```

```python
# 모델 평가
print('정답')
print(y)
print('-'*80)
print('모델이 예측한 값')
print(model.predict(x).reshape(-1))
```
